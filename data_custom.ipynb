Apres avoir analyser les données , on peut se rendre compte qu'il y'a  des colones dont le nombres  de valeurs manquantes arrivents jusqu'a 81 %.
Ces  données là, vue qu'ils sont en majorité vide , nous avons decider de les supprimés

```python
threshold = 0.5  # le seuil (50 %) de valeurs manquantes
data = data.dropna(axis=1, thresh=int(threshold * len(data)))
```

```python
data.isna().sum() /data.shape[0]*100
```

Apres avoir supprimmer les données ayant enormement de valeurs manquantes, nous allons devoir remplacer les valeurs manquantes pour nos colones restante.
Pour les colones de type object nous allons remplacer les valeurs manquant

```python
for col in  data.columns :
    if col in data.select_dtypes(include=['object']) :
        data.fillna({col: data[col].mode()[0]}, inplace=True)
    else :
        data.fillna({col: data[col].mean()}, inplace=True)
```

Generation des prix

```python
def price_generator(data):

    base_price = 300  # Prix de base imposer

    # Ajustement du prix selon la distance
    price = base_price + data['DISTANCE'] * 0.15 

    # Ajustement du prix selon le jour de la semaine
    if data['DAY_OF_WEEK'] in [5, 7]:  
        price = price + (price * 0.2) 

    # Ajustement du prix selon le mois (vacances)
    if data['MONTH'] in [6, 7, 8, 12]:  
        price = price + (price*0.5) 

    # Ajustements basés sur la compagnie aérienne
    airline = {
        'AA': np.random.uniform(0.20, 0.50), 'AS': np.random.uniform(0.20, 0.50), 'B6': np.random.uniform(0.20, 0.50), 'DL': np.random.uniform(0.20, 0.50), 'EV': np.random.uniform(0.20, 0.50), 'F9': np.random.uniform(0.20, 0.50),
        'HA': np.random.uniform(0.20, 0.50), 'MQ': np.random.uniform(0.20, 0.50), 'NK': np.random.uniform(0.20, 0.50), 'OO': np.random.uniform(0.20, 0.50), 'UA':np.random.uniform(0.20, 0.50), 'US': np.random.uniform(0.20, 0.50),
        'VX': np.random.uniform(0.20, 0.50), 'WN': np.random.uniform(0.20, 0.50)
    }
    
    for key , value in airline.items():
        if data.get('AIRLINE_'+key) == 1:
              price = price + (price*value) 
    
        
    if 500 <= data['SCHEDULED_TIME'] <= 1900:
        price = price + (price*0.25 )
    

    return price
```

```python
new_data['PRICE'] = new_data.apply(price_generator, axis=1)
```

```python
y = new_data['PRICE'] 
X = new_data.drop(['PRICE','ORIGIN_AIRPORT','DESTINATION_AIRPORT'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression , ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
models = {
    'ElasticNet': ElasticNet(),
    'DecisionTree': DecisionTreeRegressor(),
    'GradientBoosting': GradientBoostingRegressor()
}

# grilles des paramètres pour chaque modèle
param_grids = {
    'ElasticNet': {
        'alpha': [0.1, 1.0, 10.0],
        'l1_ratio': [0.1, 0.5, 0.9],
        'max_iter': [10, 50, 100]
    },
    'DecisionTree': {
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    },
    'GradientBoosting': {
        'n_estimators': [50, 100, 200],
        'learning_rate': [0.01, 0.1, 0.2],
        'max_depth': [3, 5, 7]
    }
}
```

```python
best_models = {}
for model_name, model in models.items():
    print(f"Training {model_name}...")
    grid_search = GridSearchCV(model, param_grids[model_name], cv=3,verbose=2) 
    grid_search.fit(X_train, y_train)
    
    best_models[model_name] = {
        'best_estimator': grid_search.best_estimator_,
        'best_params': grid_search.best_params_,
        'best_score': grid_search.best_score_
    }
```

```python
for model_name, result in best_models.items():
    print(f"\nModel: {model_name}")
    print(f"Best Parameters: {result['best_params']}")
    print(f"Best CV Score: {result['best_score']}")

best_model_name = max(best_models, key=lambda x: best_models[x]['best_score'])
best_model = best_models[best_model_name]['best_estimator']

print(f"\nBest Model: {best_model_name}")
print(f"Best Estimator: {best_model}")

y_pred = best_model.predict(X_test)
test_mse = mean_squared_error(y_test, y_pred)
print(f"Test MSE for Best Model: {test_mse}")
```
